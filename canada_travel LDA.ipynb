```python
#CANADA
#provinces
p = can_cities['province_name'].unique()
p = [x.lower() for x in p]
province = '|'.join(str(x) for x in p)

#provinces abv
pa = can_cities['province_id'].unique()
pa = [x.lower() for x in pa]
prov_abv = '|'.join(str(x) for x in pa)

city = '|'.join(str(x) for x in can_cities['city_ascii'].str.lower())
combined = province + '|' + city + '|' + prov_abv

#USA cities
us_city = '|'.join(str(x) for x in us_cities['city_ascii'].str.lower())
```

```python
import re
'''
    Parameter:
        search_str(str): target string that we want to check
        search_list_country(str): list of country that has been join as string pattern
        search_list_can(str): list of Canada's town that has been join as pattern
        search_list_us(str): list of USA's town that has been join as pattern
    Return:
        return_str(str): name of Country, or Canada/ USA if it's match the CAN & US city pattern, unknown if no pattern match
''' 

def can_city_search(search_str:str, search_list_country=country_pattern, search_list_can=combined, search_list_us=us_city):
    country_obj = re.search(search_list_country, search_str)
    city_obj = re.search(search_list_can, search_str)
    us_obj = re.search(search_list_us, search_str)
    if country_obj :
        return_str = search_str[country_obj.start(): country_obj.end()]
    elif city_obj:
        return_str = 'canada'
    elif us_obj:
        return_str = 'united states'
    else:
        return_str = 'unknown'
    return return_str
```

```python
#apply the can_city_search function to search the destination country
data['country'] = data['destination_en'].apply(lambda x: can_city_search(search_str=x))
data.head()
```

```python
#function to check words contains and change the country
def country_replace(regx, replacement:str):
    if type(regx) == list:
        for x in regx:
            mask = data['destination_en'].str.match(x)
            data.loc[mask, 'country'] = data.loc[mask, 'country'].replace({'unknown':replacement}, regex=True)
    else:
        mask = data['destination_en'].str.contains(regx)
        data.loc[mask, 'country'] = data.loc[mask, 'country'].replace({'unknown':replacement}, regex=True)
```

## topic modeling Latent Dirichlet Allocation

Before we implement LDA, we have to clean our purpose data by removing punctuation, numbers and make sure all of the data in english. So we will convert the data to document term matrix.

```python
#remove punctuation
data['purpose_en_clean'] = data['purpose_en'].str.replace('[^\w\s]', ' ', regex=True)
#remove number
data['purpose_en_clean'] = data['purpose_en'].str.replace('[0-9]', '', regex=True)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(stop_words='english')
purpose_cv = cv.fit_transform(data['purpose_en_clean'])
purpose_matrix = pd.DataFrame(purpose_cv.toarray(), columns=cv.get_feature_names_out())
purpose_matrix
```

While we create the document term matrix, we found there are plenty of random words and abbreviation. So we will be pulling only noun and adjectives from the purpose column.

```python
import nltk
nltk.download('averaged_perceptron_tagger')
nltk.download('wordnet')
```

```python
from nltk import word_tokenize, pos_tag
from nltk.stem import WordNetLemmatizer
```

```python
#function to pull out nouns and adj from text
def noun_adj(text):
    is_noun_adj = lambda pos: pos[:2] == 'NN' or pos[:2] == 'JJ'
    tokenized = word_tokenize(text)
    nouns_adj = [word for (word, pos) in pos_tag(tokenized) if is_noun_adj(pos)] 
    return ' '.join(nouns_adj)
```

```python
# Apply the nouns function to the transcripts to filter only on nouns
data_nouns_adj = pd.DataFrame(data['purpose_en_clean'].apply(noun_adj))
data_nouns_adj
```

```python
# Create a new document-term matrix using only nouns and adjectives, also remove common words with max_df
cvna = CountVectorizer(stop_words='english', max_df=.8)
data_cvna = cvna.fit_transform(data_nouns_adj.purpose_en_clean)
data_dtmna = pd.DataFrame(data_cvna.toarray(), columns=cvna.get_feature_names_out())
data_dtmna
```

```python
from gensim import matutils, models
import scipy.sparse
```

```python
# Create the gensim corpus
corpusna = matutils.Sparse2Corpus(scipy.sparse.csr_matrix(data_dtmna.transpose()))

# Create the vocabulary dictionary
id2wordna = dict((v, k) for k, v in cvna.vocabulary_.items())
```

```python
# Let's try 4 topics
ldana = models.LdaModel(corpus=corpusna, num_topics=4, id2word=id2wordna, passes=10)
ldana.print_topics()
```
